```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
# Load files
loss = np.load('./losses.npy')
loss_val = np.load('./losses_val.npy')
pixel_acc = np.load('./p_acc.npy', allow_pickle=True)
iou_acc = np.load('./iou_acc.npy')
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
axs = axs.flatten()
axs[0].plot(loss)
axs[0].plot(loss_val)
axs[0].set_xlabel('Iteration')
axs[0].set_title('Loss')

axs[1].plot(pixel_acc)
axs[1].set_xlabel('Iteration')
axs[1].set_title('Pixel Accuracy')
```

```python
fig, ax = plt.subplots(figsize=(9,5))
ax.plot(iou_acc)
ax.set_xlabel('Iteration')
ax.set_title('IOU Accuracy')
```
